```python

df = pd.read_excel('/default_of_credit_card_clients__courseware_version_1_21_19.xls')
```

**Número de colunas**

```python
total_colunas = df.shape[1]
print(f"Número de colunas: {total_colunas}")
```

**Número de linhas**

```python
total_linhas = df.shape[0]
print(f"Número de linhas: {total_linhas}")
```

**Tipos de características**

```python
tipos = df.dtypes
print("Tipos de características:")
print(tipos)
```

**Separação entre categóricas e numéricas**

```python
categ_colunas = df.select_dtypes(include=['object', 'category']).columns.tolist()
num_colunas = df.select_dtypes(include=['number']).columns.tolist()
print(f"Colunas categóricas: {categ_colunas}")
print(f"Colunas numéricas: {num_colunas}")
```

**Resumo dos dados numéricos**

```python
descricao_numericas = df[num_colunas].describe()
print("Resumo estatístico das características numéricas:")
print(descricao_numericas)
```

**Resumo das características categóricas**

```python
for col in categ_colunas:
    print(f"Distribuição da característica categórica '{col}':")
    print(df[col].value_counts())
    print("\n")
```

**Verificar dados faltantes**

```python
dados_faltantes = df.isnull().sum()
print("Dados faltantes por coluna:")
print(dados_faltantes[dados_faltantes > 0])
```
